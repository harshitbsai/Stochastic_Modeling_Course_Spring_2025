```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

from scipy.special import beta
def calculate_a(mu, sigma):
    a = (mu**2 * (1 - mu) / sigma**2) - mu
    return a

def calculate_b(mu, sigma):
    b = (mu * (1 - mu)**2 / sigma**2) - (1 - mu)
    return b

# Example usage:
mu = 0.5
sigma = 0.4

a = calculate_a(mu, sigma)
b = calculate_b(mu, sigma)

print(f"a = {a}")
print(f"b = {b}")

def compute_beta(a, b):
  """
  Computes the beta function for given parameters a and b.

  Args:
    a: First parameter (must be greater than 0).
    b: Second parameter (must be greater than 0).

  Returns:
    The value of the beta function.
  """
  return beta(a, b)

result = compute_beta(a, b)
print(f"Beta function for a={a} and b={b}: {result}")
```

```python
def f(x):
  return 1/(compute_beta(a,b)) *(x**(a-1))*((1-x)**(b-1))
```

```python
x = np.linspace(0, 1, 100)

# Calculate f(x) for each x value
y_values = f(x)

# Plot the Beta PDF
plt.plot(x_values, y_values, label=f"Beta PDF (a={a}, b={b})")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Beta Distribution PDF")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.special import beta

# Set up the figure size and layout
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

# Function to calculate parameter 'a'
def calculate_a(mu, sigma):
    a = (mu**2 * (1 - mu) / sigma**2) - mu
    return a

# Function to calculate parameter 'b'
def calculate_b(mu, sigma):
    b = (mu * (1 - mu)**2 / sigma**2) - (1 - mu)
    return b

# Function to compute the Beta function
def compute_beta(a, b):
    return beta(a, b)

# Function to compute the Beta PDF
def f(x, a, b):
    return 1 / (compute_beta(a, b)) * (x**(a-1)) * ((1-x)**(b-1))

# Fixed mu value
mu = 0.5

# List of sigma values
sigma_values = [0.4, 0.5 / np.sqrt(3), 0.15]


# Generate x values
x_values = np.linspace(0, 1, 100)

# Plot the Beta PDF for each sigma value
for sigma in sigma_values:
    a = calculate_a(mu, sigma)
    b = calculate_b(mu, sigma)
    y_values = f(x_values, a, b)
    plt.plot(x_values, y_values, label=f"Ïƒ = {sigma} (a={a:.2f}, b={b:.2f})")

# Add labels, title, legend, and grid
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Beta Distribution PDF for Different Ïƒ Values (Î¼ = 0.5)")
plt.legend()
plt.show()
```
